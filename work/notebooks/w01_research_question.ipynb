## 1. My lane (or freestyle) and why

*Name your lane — or say 'freestyle' and describe your own question. One short paragraph: why this one?*

```python
## 1. My Lane and Why

**Selected Lane:** Lane 1 — Content Refresh / Content Opportunity Scoring

**Why this lane:**
Websites contain thousands of pages, but SEO and content teams operate under strict time and resource constraints. They cannot manually audit or rewrite every single URL each week. Lane 1 addresses this exact capacity bottleneck: rather than relying on guesswork or static rules, we build a data-driven model to rank pages by their urgency for a refresh. By prioritizing declining, high-demand pages, we ensure editorial resources are directed where they prevent traffic loss and maximize recovery.
```

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
## 2. The Question: Decision, Action, and Cost of a Wrong Call

* **Unit of Analysis:** A single page (`page_id` / URL level).
* **The Decision:** Should an editorial/SEO team allocate 3–5 hours of specialized labor to audit and rewrite this specific page this week?
* **The Action:** Place the page into the top-priority "Content Refresh Queue" and deliver a diagnostic reason code pointing the editor to the exact weakness (e.g., outdated content, low engagement, or position decay).
* **Cost of False Positives (False Alarms):** Wasted human effort, salary budget, and opportunity cost spent modifying a page that was already performing well—potentially even breaking what made it rank in the first place.
* **Cost of False Negatives (Missed Opportunities):** Severe financial and operational damage. High-value pages (e.g., enterprise conversion pages or critical information hubs) quietly lose search visibility. Unchecked traffic decay turns potential multi-million-dollar leads away to competitors simply because nobody caught the drop in time.
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# Clone your repo into the current Colab environment
!git clone https://github.com/jhaaryan7580-dev/FlyRank-AI-.git

# Move into the notebooks folder so paths line up
%cd FlyRank-AI-/work/notebooks
```

```python
import pandas as pd
import os

# Try finding the file across common Colab directory structures
possible_paths = [
    "../../data/raw/content_refresh_anonymized.csv",
    "data/raw/content_refresh_anonymized.csv",
    "FlyRank-AI-/data/raw/content_refresh_anonymized.csv",
    "/content/FlyRank-AI-/data/raw/content_refresh_anonymized.csv"
]

csv_path = None
for path in possible_paths:
    if os.path.exists(path):
        csv_path = path
        break

if csv_path:
    df = pd.read_csv(csv_path)
    print(f"Successfully loaded dataset from: {csv_path}")
else:
    raise FileNotFoundError("Could not find content_refresh_anonymized.csv. Check repo cloning in Colab.")

total_pages = len(df)
declining_pages = (df["trend_direction"] == "down").sum()
pct_declining = (declining_pages / total_pages) * 100

total_90d_impressions = df["impressions_90d"].sum()
declining_impressions = df[df["trend_direction"] == "down"]["impressions_90d"].sum()
pct_declining_impressions = (declining_impressions / total_90d_impressions) * 100

print(f"1. Total Dataset Size: {total_pages:,} pages")
print(f"2. Declining Pages: {declining_pages:,} ({pct_declining:.1f}% of total pages)")
print(f"3. Traffic at Risk: Declining pages account for {declining_impressions:,} impressions ({pct_declining_impressions:.1f}% of all 90d impressions)")
```
